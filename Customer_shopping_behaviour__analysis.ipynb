```python
# filling missing values of review rating

''' We are filling missing values with median
value because if we use mean value it could be
effected by outliers but in case of median its
fixed(the middle most value)'''

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))

'''we also grouped the review data with category
because every category of item have differnt
review ratings and differnts medians as well. '''
```

```python
# converting columns into snake case

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns.tolist()
```

```python
# creating a new column age_group

labels = ['Young Adult','Adult','Middle Aged','senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels =labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python


# creating a new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30,
    'Every 3 Months':90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
